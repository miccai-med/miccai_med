**Description:** This model uses MobileNetV2 as a feature extractor, feeding into a multi-stage adapter that branches into two parallel Vision Transformer text encoders and a CNN head.

```python
"""
MobileFormer Training Script - 256:48:48 Ratio with x1.25 Multiplier + BatchNorm
=================================================================================
This script trains a hybrid CNN-ViT model (MobileFormer) for classifying
chest X-ray images into three classes: Cardiomegaly, Edema, and Pneumothorax.

Key Architectural Features:
- Uses timm VisionTransformer (proper ViT with patch embedding)
- VIT-1 (vit_encoder1): Full adapter output (56x56x16) with overlapping patches -> 48 features
- VIT-2 (vit_encoder2): Intermediate (28x28x32) + TwoMaxMinusMinPool2d + x1.25 -> 48 features
- CNN parallel head with 256 output features
- Fusion: CNN(256) + VIT-1(48) + VIT-2(48) = 352 features
- **NEW: BatchNorm1d layer after concat, before classifier**

CHANGES FROM 256:48:48:
- Added BatchNorm1d(352) after feature concatenation
- This normalizes the combined features before final classification

Date: 2025
"""

# ============================================================================
# CELL 1: Import Required Libraries
# ============================================================================
import os
import time
from datetime import timedelta
from pathlib import Path

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from PIL import Image
from tqdm import tqdm

import torch
import torch.nn as nn
import torch.nn.functional as F
from torch.utils.data import Dataset, DataLoader, random_split
from torchvision import transforms
from torchvision.models import mobilenet_v2, MobileNet_V2_Weights

from sklearn.metrics import (
    accuracy_score, precision_score, recall_score, f1_score,
    confusion_matrix, classification_report, roc_auc_score,
    matthews_corrcoef
)

# timm for proper VisionTransformer
from timm.models.vision_transformer import VisionTransformer

# Model visualization imports
from torchinfo import summary
from torchview import draw_graph
from IPython.display import SVG, display

# ============================================================================
# CELL 2: Configuration Class and Hyperparameters
# ============================================================================
class Config:
    """Configuration class containing all hyperparameters and paths."""
    # Paths - UPDATE THIS TO YOUR DATASET PATH
    DATA_DIR = "/kaggle/input/chest-xray-4class-100k"
    OUTPUT_DIR = "/kaggle/working"
    WEIGHTS_PATH = "/kaggle/input/mobilenetv2-weights/mobilenet_v2-b0353104.pth"
    
    # Model configuration
    MODEL_NAME = "MobileFormer-CNN-ViT-256-48-48-Mult125-BN"
    NUM_CLASSES = 3
    LABELS = ["Cardiomegaly", "Edema", "Pneumothorax"]
    
    # Training hyperparameters
    BATCH_SIZE = 8
    NUM_EPOCHS = 25
    LEARNING_RATE = 1e-4
    WEIGHT_DECAY = 1e-5
    
    # Image configuration
    IMG_SIZE = 224
    
    # Data split
    TRAIN_SPLIT = 0.8
    
    # Device configuration
    DEVICE = torch.device("cuda" if torch.cuda.is_available() else "cpu")
    
    # Random seed
    SEED = 42
    
    # Multiplier for VIT-2 path (replaces GELU)
    VIT2_MULTIPLIER = 1.25

config = Config()

def set_seed(seed):
    np.random.seed(seed)
    torch.manual_seed(seed)
    torch.cuda.manual_seed_all(seed)
    torch.backends.cudnn.deterministic = True
    torch.backends.cudnn.benchmark = False

set_seed(config.SEED)

print(f"Device: {config.DEVICE}")
print(f"Model: {config.MODEL_NAME}")
print(f"Labels: {config.LABELS}")
print(f"VIT-2 Multiplier: {config.VIT2_MULTIPLIER}")

# ============================================================================
# CELL 3: Dataset Class Definition
# ============================================================================
class ChestXRayDataset(Dataset):
    def __init__(self, df, img_dir, transform=None):
        self.df = df.reset_index(drop=True)
        self.img_dir = img_dir
        self.transform = transform
        self.labels = config.LABELS
        
    def __len__(self):
        return len(self.df)
    
    def __getitem__(self, idx):
        row = self.df.iloc[idx]
        img_path = row['new_path']
        full_path = os.path.join(self.img_dir, img_path)
        
        try:
            image = Image.open(full_path).convert('RGB')
        except Exception as e:
            print(f"Error loading {full_path}: {e}")
            image = Image.new('RGB', (config.IMG_SIZE, config.IMG_SIZE), (0, 0, 0))
        
        if self.transform:
            image = self.transform(image)
        
        labels = []
        for label in self.labels:
            val = row[label] if label in row.index else 0
            labels.append(1.0 if val == 1.0 else 0.0)
        
        labels = torch.tensor(labels, dtype=torch.float32)
        return image, labels

# ============================================================================
# CELL 4: Data Transforms (No augmentation as per user preference)
# ============================================================================
import pandas as pd
from sklearn.model_selection import train_test_split

train_transform = transforms.Compose([
    transforms.Resize((config.IMG_SIZE, config.IMG_SIZE)),
    transforms.RandomHorizontalFlip(p=0.5),
    transforms.RandomRotation(10),
    transforms.ColorJitter(brightness=0.2, contrast=0.2),
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
])

val_transform = transforms.Compose([
    transforms.Resize((config.IMG_SIZE, config.IMG_SIZE)),
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
])

# ============================================================================
# CELL 5: Custom Pooling Layer - TwoMaxMinusMinPool2d
# ============================================================================
class TwoMaxMinusMinPool2d(nn.Module):
    """
    Custom pooling layer: output = (2 * max_pool) - min_pool
    """
    def __init__(self, kernel_size, stride=None, padding=0):
        super().__init__()
        self.kernel_size = kernel_size
        self.stride = stride if stride is not None else kernel_size
        self.padding = padding

    def forward(self, x):
        max_pool = F.max_pool2d(x, kernel_size=self.kernel_size, 
                                 stride=self.stride, padding=self.padding)
        min_pool = -F.max_pool2d(-x, kernel_size=self.kernel_size, 
                                  stride=self.stride, padding=self.padding)
        return (2.0 * max_pool) - min_pool

# ============================================================================
# CELL 6: CNN-ViT Hybrid Model (256:48:48 + BatchNorm before classifier)
# ============================================================================
class CNN_ViT_Model(nn.Module):
    """
    Hybrid CNN-ViT Model - 256:48:48 ratio with x1.25 multiplier and BatchNorm.
    
    Architecture:
    - VIT-1: Full adapter output (56x56x16) with overlapping patches -> 48 features
    - VIT-2: Intermediate (28x28x32) + TwoMaxMinusMinPool2d + x1.25 -> 48 features
    - CNN: Concatenated features -> 256 features
    - Fusion: CNN(256) + VIT-1(48) + VIT-2(48) = 352 features
    - **BatchNorm1d(352) after concat**
    - Classifier head
    
    Key Changes from 256:48:48:
    - Added BatchNorm1d(352) layer after feature concatenation
    - This provides feature normalization before the final linear layer
    """
    def __init__(self, num_classes, weights_path=None, multiplier=1.25):
        super().__init__()
        
        self.multiplier = multiplier

        # Load pretrained MobileNetV2
        if weights_path and os.path.exists(weights_path):
            pretrained_model = mobilenet_v2(weights=None)
            state_dict = torch.load(weights_path, map_location="cpu")
            pretrained_model.load_state_dict(state_dict)
        else:
            pretrained_model = mobilenet_v2(weights=MobileNet_V2_Weights.IMAGENET1K_V1)
        
        num_blocks1 = len(pretrained_model.features)

        # Freeze first half of MobileNetV2
        for param in pretrained_model.features[:num_blocks1 // 2].parameters():
            param.requires_grad = False

        # Split MobileNetV2 into blocks
        self.mobilenet_feature_blocks_1 = nn.Sequential(*pretrained_model.features[:num_blocks1 - 3])
        self.mobilenet_block_2nd_last = pretrained_model.features[num_blocks1 - 3]
        self.mobilenet_block_mid = pretrained_model.features[num_blocks1 - 2]
        self.mobilenet_block_last = pretrained_model.features[num_blocks1 - 1]

        # Output channel sizes
        ch_2nd_last = self.mobilenet_block_2nd_last.out_channels  # 160
        ch_last = self.mobilenet_block_last.out_channels          # 1280
        ch_mid = self.mobilenet_block_mid.out_channels            # 320
        ch_concat = ch_last + ch_2nd_last                         # 1440
        
        # CNN to ViT adapter
        adapter_layers = [
            nn.Conv2d(ch_mid, 64, kernel_size=3, padding=1), nn.BatchNorm2d(64), nn.ReLU(inplace=True),
            nn.Upsample(size=(14, 14), mode="bilinear", align_corners=False),
            nn.Conv2d(64, 32, kernel_size=3, padding=1), nn.BatchNorm2d(32), nn.ReLU(inplace=True),
            nn.Upsample(size=(28, 28), mode="bilinear", align_corners=False),
            nn.Conv2d(32, 16, kernel_size=3, padding=1), nn.BatchNorm2d(16), nn.GELU(),
            nn.Upsample(size=(56, 56), mode="bilinear", align_corners=False),
        ]
        self.cnn_to_vit_adapter = nn.Sequential(*adapter_layers)

        # VIT-1: 56x56 feature maps with overlapping patches -> 48 features
        self.vit_encoder1 = VisionTransformer(
            img_size=56,
            patch_size=8,
            in_chans=16,
            num_classes=0,
            embed_dim=48,  # VIT-1: 48 features
            depth=2,
            num_heads=4,
            mlp_ratio=2,
            qkv_bias=True,
            norm_layer=nn.LayerNorm,
        )
        self.vit_encoder1.head = nn.Identity()
        
        # Enable overlapping patches by changing patch projection stride
        old_proj = self.vit_encoder1.patch_embed.proj
        self.vit_encoder1.patch_embed.proj = nn.Conv2d(
            old_proj.in_channels,
            old_proj.out_channels,
            kernel_size=8,
            stride=4,
            bias=old_proj.bias is not None,
        )
        
        # Infer number of patches produced by the new stride
        img_h, img_w = self.vit_encoder1.patch_embed.img_size
        dummy = torch.zeros(1, old_proj.in_channels, img_h, img_w)
        
        with torch.no_grad():
            _, _, h, w = self.vit_encoder1.patch_embed.proj(dummy).shape
        
        num_patches = h * w
        self.vit_encoder1.patch_embed.num_patches = num_patches
        
        # Reset positional embeddings to match new token count
        self.vit_encoder1.pos_embed = nn.Parameter(
            torch.zeros(1, num_patches + 1, self.vit_encoder1.embed_dim)
        )
        nn.init.trunc_normal_(self.vit_encoder1.pos_embed, std=0.02)

        # TwoMaxMinusMinPool2d for VIT-2 path
        self.twomax_minus_min_pool = TwoMaxMinusMinPool2d(kernel_size=2, stride=1)
        # NOTE: Using multiplier instead of GELU

        # VIT-2: 28x28 feature maps (after TwoMaxMinusMinPool2d) -> 48 features
        self.vit_encoder2 = VisionTransformer(
            img_size=28, patch_size=7, in_chans=32,
            num_classes=0, embed_dim=48,  # VIT-2: 48 features
            depth=2, num_heads=4, mlp_ratio=2,
            qkv_bias=True, norm_layer=nn.LayerNorm,
        )
        self.vit_encoder2.head = nn.Identity()

        # CNN parallel head (from concatenated features)
        self.cnn_parallel_head = nn.Sequential(
            nn.Conv2d(ch_concat, 256, kernel_size=1), nn.ReLU(inplace=True),
            nn.Conv2d(256, 256, kernel_size=3, padding=0), nn.BatchNorm2d(256), nn.ReLU(inplace=True),
            nn.AdaptiveAvgPool2d(1), nn.Flatten(),
        )

        # Total fusion features: CNN(256) + VIT-1(48) + VIT-2(48) = 352
        fusion_features = 256 + 48 + 48  # 352
        
        # NEW: BatchNorm1d layer after concatenation, before classifier
        self.fusion_bn = nn.BatchNorm1d(fusion_features)
        
        # Final classifier
        self.classifier_head = nn.Linear(fusion_features, num_classes)

    def forward(self, x):
        features_1 = self.mobilenet_feature_blocks_1(x)
        features_2nd_last = self.mobilenet_block_2nd_last(features_1)
        features_mid = self.mobilenet_block_mid(features_2nd_last)
        features_last = self.mobilenet_block_last(features_mid)

        # Concatenate for CNN parallel path
        concatenated_cnn_features = torch.cat([features_last, features_2nd_last], dim=1)

        # Adapter path
        adapter_x = features_mid
        intermediate = None
        
        for i, layer in enumerate(self.cnn_to_vit_adapter):
            adapter_x = layer(adapter_x)
            # Capture intermediate at 28x28x32 stage (before last 4 layers)
            if i == len(self.cnn_to_vit_adapter) - 5:
                intermediate = adapter_x
        
        vit_input1 = adapter_x      # Full adapter output: 56x56x16
        vit_input2 = intermediate   # Intermediate: 28x28x32
        
        # VIT-1: Full adapter output -> 48 features
        vit_features1 = self.vit_encoder1(vit_input1)  # 48
        
        # VIT-2: Intermediate + TwoMaxMinusMinPool2d + x1.25 multiplier -> 48 features
        vit_input2 = self.twomax_minus_min_pool(vit_input2)
        vit_input2 = vit_input2 * self.multiplier  # x1.25 multiplier instead of GELU
        vit_input2 = F.interpolate(vit_input2, size=(28, 28), mode="bilinear", align_corners=False)
        vit_features2 = self.vit_encoder2(vit_input2)  # 48
                
        # CNN parallel branch -> 256 features
        cnn_features = self.cnn_parallel_head(concatenated_cnn_features)  # 256

        # Concatenate: CNN(256) + VIT-1(48) + VIT-2(48) = 352
        combined = torch.cat([vit_features1, vit_features2, cnn_features], dim=1)
        
        # NEW: Apply BatchNorm1d before classifier
        combined = self.fusion_bn(combined)
        
        return self.classifier_head(combined)

# ============================================================================
# CELL 7: Training and Validation Functions
# ============================================================================
def train_epoch(model, loader, criterion, optimizer, device):
    model.train()
    running_loss = 0.0
    all_preds = []
    all_labels = []
    
    pbar = tqdm(loader, desc="Training")
    for images, labels in pbar:
        images = images.to(device)
        labels = labels.to(device)
        
        optimizer.zero_grad()
        outputs = model(images)
        loss = criterion(outputs, labels)
        
        loss.backward()
        optimizer.step()
        
        running_loss += loss.item()
        
        preds = torch.sigmoid(outputs).detach().cpu().numpy()
        all_preds.extend(preds)
        all_labels.extend(labels.cpu().numpy())
        
        pbar.set_postfix({'loss': f'{loss.item():.4f}'})
    
    epoch_loss = running_loss / len(loader)
    all_preds = np.array(all_preds)
    all_labels = np.array(all_labels)
    binary_preds = (all_preds > 0.5).astype(int)
    accuracy = (binary_preds == all_labels).mean()
    
    return epoch_loss, accuracy

def validate_epoch(model, loader, criterion, device):
    model.eval()
    running_loss = 0.0
    all_preds = []
    all_probs = []
    all_labels = []
    
    with torch.no_grad():
        pbar = tqdm(loader, desc="Validation")
        for images, labels in pbar:
            images = images.to(device)
            labels = labels.to(device)
            
            outputs = model(images)
            loss = criterion(outputs, labels)
            
            running_loss += loss.item()
            
            probs = torch.sigmoid(outputs).cpu().numpy()
            all_probs.extend(probs)
            all_preds.extend((probs > 0.5).astype(int))
            all_labels.extend(labels.cpu().numpy())
    
    epoch_loss = running_loss / len(loader)
    all_preds = np.array(all_preds)
    all_labels = np.array(all_labels)
    all_probs = np.array(all_probs)
    accuracy = (all_preds == all_labels).mean()
    
    return epoch_loss, accuracy, all_preds, all_probs, all_labels

# ============================================================================
# CELL 8: Main Training Pipeline
# ============================================================================
def main():
    print("="*80)
    print("MobileFormer Training - CNN(256) + VIT-1(48) + VIT-2(48) = 352 features")
    print("256:48:48 Ratio with x1.25 Multiplier + BatchNorm before classifier")
    print("VIT-2 (28x28 input) uses TwoMaxMinusMinPool2d + x1.25 multiplier")
    print("NEW: BatchNorm1d(352) after feature fusion")
    print("="*80)
    
    # Data Loading
    print("\n" + "="*60)
    print("LOADING DATA")
    print("="*60)
    
    metadata_path = os.path.join(config.DATA_DIR, "metadata.csv")
    
    if not os.path.exists(metadata_path):
        print(f"\nWARNING: Metadata file '{metadata_path}' not found!")
        return
    
    df = pd.read_csv(metadata_path)
    print(f"Total samples: {len(df)}")
    
    print("\nLabel distribution:")
    for label in config.LABELS:
        if label in df.columns:
            count = (df[label] == 1.0).sum()
            print(f"  {label}: {count} ({count/len(df)*100:.2f}%)")
    
    train_df, val_df = train_test_split(
        df, test_size=(1 - config.TRAIN_SPLIT), random_state=config.SEED
    )
    
    print(f"\nTrain samples: {len(train_df)}")
    print(f"Validation samples: {len(val_df)}")
    
    train_dataset = ChestXRayDataset(train_df, config.DATA_DIR, train_transform)
    val_dataset = ChestXRayDataset(val_df, config.DATA_DIR, val_transform)
    
    train_loader = DataLoader(train_dataset, batch_size=config.BATCH_SIZE, 
                              shuffle=True, num_workers=2, pin_memory=True)
    val_loader = DataLoader(val_dataset, batch_size=config.BATCH_SIZE, 
                            shuffle=False, num_workers=2, pin_memory=True)
    
    # Model Initialization
    print("\n" + "-"*40)
    print("Initializing Model...")
    print(f"VIT-2 Path: TwoMaxMinusMinPool2d + x{config.VIT2_MULTIPLIER} multiplier")
    print(f"Feature Fusion: CNN(256) + VIT-1(48) + VIT-2(48) = 352")
    print("NEW: BatchNorm1d(352) before classifier")
    print("-"*40)
    
    model = CNN_ViT_Model(
        num_classes=config.NUM_CLASSES,
        weights_path=config.WEIGHTS_PATH,
        multiplier=config.VIT2_MULTIPLIER
    )
    model = model.to(config.DEVICE)
    
    total_params = sum(p.numel() for p in model.parameters())
    trainable_params = sum(p.numel() for p in model.parameters() if p.requires_grad)
    print(f"\nTotal parameters: {total_params:,}")
    print(f"Trainable parameters: {trainable_params:,}")
    
    # Model Summary
    print("\n" + "="*60)
    print("MODEL SUMMARY")
    print("="*60)
    
    summary(model, input_size=(1, 3, 224, 224),
            col_names=["input_size", "output_size", "num_params", "trainable"])
    
    # Model Visualization
    print("\n" + "="*60)
    print("GENERATING MODEL VISUALIZATION")
    print("="*60)
    
    dummy_input = torch.randn(1, 3, 224, 224, device=config.DEVICE)
    
    model_graph = draw_graph(
        model, input_data=dummy_input, expand_nested=True,
        graph_name="CNN_ViT_Model_256_48_48_Mult125_BN", save_graph=False
    )
    
    svg_bytes = model_graph.visual_graph.pipe(format="svg")
    jpg_bytes = model_graph.visual_graph.pipe(format="jpg")
    
    with open(os.path.join(config.OUTPUT_DIR, "CNN_ViT_Model_256_48_48_Mult125_BN.svg"), "wb") as f:
        f.write(svg_bytes)
    with open(os.path.join(config.OUTPUT_DIR, "CNN_ViT_Model_256_48_48_Mult125_BN.jpg"), "wb") as f:
        f.write(jpg_bytes)
    
    print("Saved: CNN_ViT_Model_256_48_48_Mult125_BN.svg, CNN_ViT_Model_256_48_48_Mult125_BN.jpg")
    display(SVG(svg_bytes))
    
    criterion = nn.BCEWithLogitsLoss()
    optimizer = torch.optim.AdamW(model.parameters(), lr=config.LEARNING_RATE, 
                                   weight_decay=config.WEIGHT_DECAY)
    scheduler = torch.optim.lr_scheduler.CosineAnnealingLR(
        optimizer, T_max=config.NUM_EPOCHS, eta_min=1e-6
    )
    
    # Training Loop
    print("\n" + "="*60)
    print("TRAINING")
    print("="*60)
    
    history = {'train_loss': [], 'train_acc': [], 'val_loss': [], 'val_acc': []}
    best_val_loss = float('inf')
    best_model_state = None
    
    for epoch in range(config.NUM_EPOCHS):
        print(f"\nEpoch {epoch+1}/{config.NUM_EPOCHS}")
        print("-"*40)
        
        train_loss, train_acc = train_epoch(model, train_loader, criterion, optimizer, config.DEVICE)
        val_loss, val_acc, val_preds, val_probs, val_labels = validate_epoch(
            model, val_loader, criterion, config.DEVICE
        )
        
        scheduler.step()
        
        history['train_loss'].append(train_loss)
        history['train_acc'].append(train_acc)
        history['val_loss'].append(val_loss)
        history['val_acc'].append(val_acc)
        
        print(f"Train Loss: {train_loss:.4f} | Train Acc: {train_acc:.4f}")
        print(f"Val Loss: {val_loss:.4f} | Val Acc: {val_acc:.4f}")
        print(f"LR: {scheduler.get_last_lr()[0]:.6f}")
        
        if val_loss < best_val_loss:
            best_val_loss = val_loss
            best_model_state = model.state_dict().copy()
            print("Best model saved!")
    
    model.load_state_dict(best_model_state)
    
    model_path = os.path.join(config.OUTPUT_DIR, "mobileformer_256_48_48_mult125_bn_model.pth")
    torch.save(best_model_state, model_path)
    print(f"\nModel saved to: {model_path}")
    
    # Final Evaluation
    print("\n" + "="*60)
    print("FINAL EVALUATION")
    print("="*60)
    
    val_loss, val_acc, all_preds, all_probs, all_labels = validate_epoch(
        model, val_loader, criterion, config.DEVICE
    )
    
    print(f"Final Validation Loss: {val_loss:.4f}")
    print(f"Final Validation Accuracy: {val_acc:.4f}")
    
    # Plot Training Curves
    fig, axes = plt.subplots(1, 2, figsize=(14, 5))
    
    axes[0].plot(history['train_loss'], label='Train Loss', marker='o')
    axes[0].plot(history['val_loss'], label='Val Loss', marker='s')
    axes[0].set_xlabel('Epoch')
    axes[0].set_ylabel('Loss')
    axes[0].set_title('Training & Validation Loss')
    axes[0].legend()
    axes[0].grid(True, alpha=0.3)
    
    axes[1].plot(history['train_acc'], label='Train Accuracy', marker='o')
    axes[1].plot(history['val_acc'], label='Val Accuracy', marker='s')
    axes[1].set_xlabel('Epoch')
    axes[1].set_ylabel('Accuracy')
    axes[1].set_title('Training & Validation Accuracy')
    axes[1].legend()
    axes[1].grid(True, alpha=0.3)
    
    plt.tight_layout()
    plt.savefig(os.path.join(config.OUTPUT_DIR, 'training_curves_256_48_48_mult125_bn.png'), dpi=300)
    plt.show()
    
    # Confusion Matrices
    fig, axes = plt.subplots(1, 3, figsize=(15, 5))
    
    for i, label in enumerate(config.LABELS):
        class_preds = all_preds[:, i]
        class_labels = all_labels[:, i]
        cm = confusion_matrix(class_labels, class_preds)
        
        sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
                    xticklabels=['Negative', 'Positive'],
                    yticklabels=['Negative', 'Positive'],
                    ax=axes[i], annot_kws={'size': 14})
        axes[i].set_xlabel('Predicted')
        axes[i].set_ylabel('Actual')
        axes[i].set_title(f'{label}')
    
    plt.tight_layout()
    plt.savefig(os.path.join(config.OUTPUT_DIR, 'confusion_matrices_256_48_48_mult125_bn.png'), dpi=300)
    plt.show()
    
    # Classification Reports
    print("\n" + "="*60)
    print("CLASSIFICATION REPORTS")
    print("="*60)
    
    for i, label in enumerate(config.LABELS):
        print(f"\n{label}:")
        print(classification_report(
            all_labels[:, i], all_preds[:, i], 
            target_names=['Negative', 'Positive'], digits=4
        ))
    
    print("\n" + "="*60)
    print("TRAINING COMPLETE!")
    print("="*60)

if __name__ == "__main__":
    main()
```
